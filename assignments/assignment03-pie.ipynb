```python
# Importing the libraries needed for my assignment:
import pandas as pd         # importing pandas to work with datasets.
import matplotlib.pyplot as plt    # matpotlib to create the pie chart.
import numpy as np                  # using for the array in building my Chart.
# pd.options.mode.chained_assignment = None  # Removing warning message. default='warn'
```

```python
# Uploading the file to my DataFrame using Pandas, from the given url.
df = pd.read_csv("https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download")
```

```python
# Checking if the file is uploaded in the DataFrame.
df
```

```python
# Get the row labels and column labels using .index and .columns.
df.index
df.columns
```

_For my assignment I need to use only data from the Email column._

```python
# Creating a subset of DataFrame to work with only emails.
df_email = df[['Email']]
df_email
```

```python
# Adding a new column with the domain name only using str.split. 
df_email['domain'] = df_email['Email'].str.split('@').str[1]
df_email
```

```python
# Number of unique domains in the DataFrame.
df_email['domain'].value_counts()
```

```python
# Get domain counts
domain_counts = df_email['domain'].value_counts()
# Convert to NumPy array
domain_array = domain_counts.values
# Checking the array.
domain_array
```

### Creating the Pie Chart.

***

```python
# Crating labels from the unique values obtained by the .value_counts we run above.
domain_labels = 'example.org', 'example.com', 'example.net'

# I will use explode to separate one slice of the pie chart from the others
myexplode = [0.2, 0, 0]

# Set figure size.
plt.figure(figsize=(8,8))

# Plotting the PieChart.
plt.pie(domain_array, labels=domain_labels, explode = myexplode, shadow = True, autopct='%1.1f%%',
       pctdistance=1.2, labeldistance=.3)

# Adding a legend to the chart.
plt.legend()

# Adding a title and formatting it. 
plt.title('Domains', fontweight='bold', fontsize=18, color='navy', bbox={'facecolor':'0.8', 'pad':5})

# Show the Chart.
plt.show()
```
