important: using 'ncols' and 'nrows'

```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
num_cols = 3
num_rows = 2
fig, axs = plt.subplots(ncols=num_cols, nrows=num_rows, figsize=(8,4))
axs[0,1].plot([1,2,3],[3,2,1])
```

```python
# different way - using for loop
num_cols = 3
num_rows = 2
fig, axs = plt.subplots(ncols=num_cols, nrows=num_rows, figsize=(8,4))
for row in range(num_rows):
    for col in range(num_cols):
        axs[row, col].annotate(f"plot[{row},{col}]", (0.5, 0.5), ha="center")
plt.show()
```

### Example

```python
data = [
    [3,5,4,9,1,6,5,7],
    [9,8,7,6,5,4,3,2],
    [1,2,3,4,5,6,7,8],
    [5,5,5,5,5,5,5,5]
] 

x = 0.5 + np.arange(8)
```

```python
# using the same code as above with little change
num_cols = 2
num_rows = 2
fig, axs = plt.subplots(ncols=num_cols, nrows=num_rows, figsize=(8,4))
count = 0
for row in range(num_rows):
    for col in range(num_cols):     
        axs[row,col].plot(x,data[count])
        count += 1
                              
plt.show()
```
